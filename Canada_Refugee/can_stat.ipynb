```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('can_stat.csv')
```

### Q1. From which countries has Canada admitted the highest number of refugees?

```python
top_refugee_country = df.groupby('Country-of-origin')['UNHCR-refugees'].sum().reset_index()
top_refugee_country = top_refugee_country.sort_values(by = 'UNHCR-refugees', ascending = False).head(10)
```

```python
plt.figure(figsize = (12,6))
sns.barplot(data = top_refugee_country, x = 'Country-of-origin', y = 'UNHCR-refugees', palette = 'mako')
plt.title('Top countries of origin for refugees')
plt.xticks(rotation = 45)
plt.show()
```

### Q2. What are the total number of resettled refugees in Canada per year?

```python
refugee_per_year = df.groupby('Year')['UNHCR-refugees'].sum().reset_index()
refugee_per_year = refugee_per_year.sort_values(by = 'Year')
refugee_per_year
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(data = refugee_per_year, x = 'Year', y = 'UNHCR-refugees', marker = 'o')
plt.title('Total number of refugees per year')
plt.xticks(rotation = 45)
plt.show()
```

### Q3. What are the countries of origin for the majority of asylum claims made in Canada?

```python
top_asylum_seeker_country = df.groupby('Country-of-origin')['Asylum-seekers'].sum().reset_index()
top_asylum_seeker_country = top_asylum_seeker_country.sort_values(by = 'Asylum-seekers', ascending = False).head(10)


plt.figure(figsize = (12,6))
sns.barplot(data = top_asylum_seeker_country, x = 'Country-of-origin', y = 'Asylum-seekers', palette = 'mako')
plt.title('Top countries of origin for Asylum seekers')
plt.xticks(rotation = 45)
plt.show()
```

### Q4. What is the total number of asylum claims made in Canada every year?

```python
as_seeker_per_year = df.groupby('Year')['Asylum-seekers'].sum().reset_index()
as_seeker_per_year = as_seeker_per_year.sort_values(by = 'Year')
as_seeker_per_year
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(data = as_seeker_per_year, x = 'Year', y = 'Asylum-seekers', marker = 'o')
plt.title('Total number of asylum seeker per year')
plt.xticks(rotation = 45)
plt.show()
```

### Q5. What are the general trends in refugee and asylum statistics from 2012-2022?

```python
top_country = df.groupby('Country-of-origin')['total-count'].sum().reset_index()
top_country = top_country.sort_values(by = 'total-count', ascending = False).head(10)
top_country
```

```python
plt.figure(figsize = (12,6))
sns.barplot(data = top_country, x = 'Country-of-origin', y = 'total-count', palette = 'mako')
plt.title('Top countries of origin for refugees and asylum seekers')
plt.xticks(rotation = 45)
plt.show()
```

```python
yearly_counts = df.groupby('Year')[['UNHCR-refugees','Asylum-seekers']].sum().reset_index()
yearly_counts_melted = pd.melt(yearly_counts, id_vars = ['Year'], var_name = 'Category', value_name = 'count')
```

```python
plt.figure(figsize = (12,6))
sns.barplot(data = yearly_counts_melted, x = 'Year', y = 'count',hue = 'Category', palette = 'mako')
plt.title('Total refugees and aslum seekers per year')
plt.xticks(rotation = 45)
plt.show()
```
