```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('DSData.csv', encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status' , 'unnamed1'], axis = 1, inplace = True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
df.info()
```

```python
df['Amount']=df['Amount'].astype('int')
df.head()
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y= 'Amount', data = sales_gen)
```

```python
#More female buyers than male.
```

```python
ax = sns.countplot(data = df, x = 'Age Group',hue = 'Gender')
for bars in ax.containers:
     ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x='Age Group', y = 'Amount', data = sales_age)
```

```python
# Most of the buyers are females of age group 26-35
```

```python
ax = sns.countplot(data = df, x = 'State')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
     ax.bar_label(bars)
```

```python
sales_state = df.groupby(['State'], as_index  = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x='State', y = 'Orders')
```

```python
# Most of the orders are from UP, Maharastra, and Karnataka.
```

```python
sales_state = df.groupby(['State'], as_index  = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x='State', y = 'Amount')
```

```python
# Highest amount of purshase is from UP, Maharastra, and Karnataka which corellates with the orders from prior graph.
# however the amount of purchase is more of Haryana than Kerela, despite Kerela having higher no. of orders.
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(5,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_status = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(data = sales_status, x = 'Marital_Status', y='Amount', hue = 'Gender')
```

```python
#More buyers are married women.
```

```python
df.groupby(['Occupation'],as_index=False).count()
```

```python
ax = sns.countplot(data = df, x = 'Occupation')
sns.set(rc={'figure.figsize':(21,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data = sales_occ, x = 'Occupation', y = 'Amount')
```

```python
#Most of the buyers are working in  IT sector, Healthcare and Aviation.
```

```python

sns.set(rc={'figure.figsize':(21,6)})
sns.countplot(data = df, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(22,6)})

sales_product= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.barplot(data = sales_product, x = 'Product_Category', y = 'Amount')
```

```python
#Most sold products are Food, Clothing and Apparel, and Electronics
```

```python
sns.set(rc={'figure.figsize':(22,6)})

sales_product_id= df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.barplot(data = sales_product_id, x = 'Product_ID', y = 'Orders')
```

```python
#Married women of age group 26-35 from UP, Maharastra and Karnataka working in IT, Healthcare
#and Aviation are most likely to buy products from Food, Clothing and Apparel, and Electronics Category.
```
