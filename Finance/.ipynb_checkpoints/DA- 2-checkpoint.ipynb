```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Financials.csv', encoding='unicode_escape')
```

```python
df.duplicated().sum()
```

```python
df.columns = df.columns.str.strip()
```

```python
dollar = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts', 'Sales', 'COGS', 'Profit']
for columns in dollar:
    df[columns] = df[columns].str.replace('$', '')
```

```python
df['Discounts'] = df['Discounts'].replace('-', '0',regex = True)
```

```python
for columns in dollar:
    df[columns] = df[columns].replace(',', '', regex = True)
```

```python
df
```

```python
cols = ['Units Sold', 'Manufacturing Price','Sale Price', 'Gross Sales', 'Discounts','Sales','COGS']
for columns in cols:
    df[columns] = df[columns].astype(float)
```

```python
(df['Profit'] == '-').sum()
```

```python
df = df[df['Profit'] != '-']
df['Profit'] = df['Profit'].str.strip()
df['Profit'] = df['Profit'].replace(r'[()]','', regex = True)
df['Profit'] = df['Profit'].astype(float)
```

```python
df.head()
```

```python
df.dtypes                                    
```

```python
df['Date']= pd.to_datetime(df['Date'])
```

```python
df = df.drop(['Month Number', 'Month Name', 'Year'],axis = 1)
```

## Country wise analysis

```python
country_grouped_df = df.groupby('Country').agg({'Sales':'sum', 'Profit': 'sum','Units Sold': 'sum','COGS': 'sum','Discounts':'sum'}).reset_index()
```

```python
plt.figure(figsize = (10,6))
plt.title('Total units sold in each Country')
bars = plt.bar('Country', 'Units Sold',data = country_grouped_df, width = 0.4)
plt.xlabel('Country')
plt.ylabel('Units Sold')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, round(height, 2),
             ha='center', va='bottom')
```

```python
plt.figure(figsize = (10,6))
plt.title('Total sales in each Country')
bars = plt.bar('Country', 'Sales',data = country_grouped_df, width = 0.4)
plt.xlabel('Country')
plt.ylabel('Sales')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, round(height, 2),
             ha='center', va='bottom')
```

### Highest sales has been in USA and Canada, lowest has been in Mexico

```python
bar_width = 0.3
plt.figure(figsize = (10,6))
plt.title('Total sales and COGS in each Country')
x = np.arange(len(country_grouped_df['Country']))
plt.bar(x-bar_width, country_grouped_df['Sales'], bar_width, label = 'Sales')
plt.bar(x, country_grouped_df['COGS'], bar_width, label = 'COGS')

plt.xticks(x, country_grouped_df['Country'])
plt.xlabel('Country')
plt.ylabel('Values')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
```

### The Company is in profit as the Sales is more than COGS in every country.

```python
plt.title('Country wise discount:')
plt.pie(country_grouped_df["Discounts"], labels = country_grouped_df["Country"], autopct = '%1.1f%%')
plt.show()
```

```python
plt.title('Country wise profit:')
plt.pie(country_grouped_df["Profit"], labels = country_grouped_df["Country"], autopct = '%1.1f%%')
plt.show()
```

### Company at France make the most profit despit Canada making the most sales, the difference between Sales price and COGS is higher in France.

## Product wise analysis

```python
product_grouped_df = df.groupby('Product').agg({'Sales':'sum', 'Profit': 'sum','Units Sold': 'sum','COGS': 'sum','Discounts':'sum'}).reset_index()
```

```python
plt.figure(figsize = (10,6))
plt.title('Total units sold of each product')
bars = plt.bar('Product', 'Units Sold',data = product_grouped_df, width = 0.4)
plt.xlabel('Product')
plt.ylabel('Units Sold')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, round(height, 2),
             ha='center', va='bottom')
```

### Maximum sold product is Paseo.

```python
plt.title('Product wise profit')
plt.pie(product_grouped_df['Profit'],labels = product_grouped_df['Product'], autopct = '%1.1f%%')
plt.show()
```

### Higest profit is from Paseo and lowest is from Montana.

```python
plt.title('Product wise discount')
plt.pie(product_grouped_df['Discounts'],labels = product_grouped_df['Product'], autopct = '%1.1f%%')
plt.show()
```

```python
bar_width = 0.3
plt.figure(figsize = (10,6))
plt.title('Sales and COGS of products')
x = np.arange(len(product_grouped_df['Product']))
plt.bar(x-bar_width, product_grouped_df['Sales'], bar_width, label = 'Sales')
plt.bar(x, product_grouped_df['COGS'], bar_width, label = 'COGS')

plt.xticks(x, product_grouped_df['Product'])
plt.xlabel('Product')
plt.ylabel('Values')
plt.legend(loc='upper right')
```

### Sales and COGS of Paseo has the most difference.

## Segment wise analysis

```python
segment_grouped_df = df.groupby('Segment').agg({'Discounts':'sum', 'Gross Sales': 'sum','Sales':'sum', 'Profit': 'sum','Units Sold': 'sum','COGS': 'sum'}).reset_index()
```

```python
plt.figure(figsize = (10,6))
plt.title('Total units sold by each segment')
bars = plt.bar('Segment', 'Units Sold',data =segment_grouped_df, width = 0.4)
plt.xlabel('Segment')
plt.ylabel('Units Sold')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, round(height, 2),
             ha='center', va='bottom')
```

### Most units are sold by Government Segment and least in Small Business

```python
plt.title('Segment wise Discounts:')
plt.pie(segment_grouped_df['Discounts'], labels = segment_grouped_df["Segment"], autopct = '%1.1f%%')
plt.show()
```

```python
plt.title('Segment wise profit:')
plt.pie(segment_grouped_df["Profit"], labels = segment_grouped_df["Segment"], autopct = '%1.1f%%')
plt.show()
```

### Government sector gives the most profit and Midmarket gives the least.

```python
bar_width = 0.3
plt.figure(figsize = (10,6))
plt.title('Sales and COGS of Segment')
x = np.arange(len(segment_grouped_df['Segment']))
plt.bar(x-bar_width, segment_grouped_df['Sales'], bar_width, label = 'Sales')
plt.bar(x, segment_grouped_df['COGS'], bar_width, label = 'COGS')

plt.xticks(x, segment_grouped_df['Segment'])
plt.xlabel('Product')
plt.ylabel('Values')
plt.legend(loc='upper right')
```

### Enterprise segment should allocate Discounts wisely.
