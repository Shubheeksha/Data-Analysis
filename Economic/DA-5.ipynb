```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('sales.csv', encoding='unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

### Dropping unnecessary columns


```python
df.drop(columns=['index','Column1'], inplace = True)
```

```python
df.dropna(inplace = True)
```

### Creating a profit column

```python
df['Profit'] = df['Revenue'] - df['Cost']
```

```python
df.head()
```

```python
df.describe()
```

## Profit Distribution.

```python
df['Profit'].hist()
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.title('Profit Distribution')
plt.show()
```

```python
gender_counts = df['Customer Gender'].value_counts()
gender_counts.plot(kind = 'pie', autopct = '%1.1f%%')
```

### Majority of the customers are male.

```python
age_quantity = df.groupby('Customer Age')['Quantity'].sum().sort_values(ascending = False).head(10).plot(kind = 'pie', y = 'Qunatity', autopct = '%1.1f%%')
plt.title('Top Age by Quantity Distribution')
plt.ylabel('') 
plt.show()
```

```python
age_quantity = df.groupby('Customer Age')['Quantity'].sum().sort_values(ascending = False).tail(10).plot(kind = 'pie', y = 'Qunatity', autopct = '%1.1f%%')
plt.title('Lowest Age by Quantity Distribution')
plt.ylabel('') 
plt.show()
```

### Most products are purchased by customer of age range 28-40 and least products are consumed by customer of age range 73-87.

## Product wise Analysis

```python
product_quantity = df.groupby('Product Category')['Quantity'].sum().reset_index()
sns.barplot(x='Product Category',y= 'Quantity', data = product_quantity,  width = 0.4)
plt.title('Quantity of product category')
```

### Top purchased category is Accessories.

```python
product_profit = df.groupby('Product Category')['Profit'].sum().reset_index()
sns.barplot(x='Product Category', y='Profit',  data=product_profit,  width = 0.4, errorbar=None)
plt.title('Profit by Product Categories')
```

```python
product_profit = df.groupby(['Product Category', 'Customer Gender'])['Profit'].sum().reset_index()
sns.barplot(x='Product Category', y='Profit', hue='Customer Gender', data=product_profit, errorbar=None)
plt.title('Categories profit by Gender')
```

### Most profit is obtained from Accessories Category which has more male customer. 

### Though bike is the second most purchased category, it gives the least profit. 

```python
country_product_quantity = df.groupby(['Country', 'Product Category'])['Quantity'].sum().reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='Quantity', hue='Product Category', data=country_product_quantity, palette='viridis')

plt.title('Top selling product catergory  by Country')
plt.xlabel('Country')
plt.ylabel('Total Quantity')

plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### All categories of product are most purchases in US. 

```python
category_quantity = df.groupby('Sub Category')['Quantity'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Sub Category',y= 'Quantity', data = category_quantity,  width = 0.4)
plt.xticks(rotation = 45)
plt.title('Products Quantity')
plt.show()
```

```python
product_profit = df.groupby(['Product Category', 'Sub Category'])['Profit'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Product Category', y='Profit', hue='Sub Category', data=product_profit, palette = 'Paired', errorbar=None)
plt.legend(title='Product Category', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.title('Profit by Subcategories')
```

### Most profit is obtained by Accessories from Helmet, Bikes from Mountain Bikes and Clothing from Jerseys.

```python
product_age = df.groupby(['Product Category', 'Customer Gender'])['Customer Age'].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.barplot(x='Product Category', y='Customer Age', hue = 'Customer Gender', data=product_age, errorbar=None)
plt.legend(title='Product Category', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.title('Average age of customer purchasing products by category')
```

```python
country_quantity = df.groupby('Country')['Quantity'].sum().reset_index()
sns.barplot(x='Country', y='Quantity', data=country_quantity, errorbar=None, width = 0.4)
```

```python
country_profit = df.groupby('Country')['Profit'].sum().reset_index()
sns.barplot(x='Country', y='Profit', data=country_profit, errorbar=None, width = 0.4)
```

### Germany makes the most profit despite having less purchased quantity.

```python
product_profit = df.groupby(['Country', 'Customer Gender'])['Quantity'].sum().reset_index()
sns.barplot(x='Country', y='Quantity', hue='Customer Gender', data=product_profit, errorbar=None)
plt.title('Gender in countries')
```

```python
country_cost = df.groupby(['Country', 'Sub Category'])['Cost'].sum().reset_index()

plt.figure(figsize=(13, 8))
sns.barplot(x='Sub Category', y='Cost', hue='Country', data=country_cost, palette='viridis')

plt.title('Cost of products in each country')
plt.xlabel('Country')
plt.ylabel('Cost')

plt.legend(title='Country', bbox_to_anchor=(1.0, 1.2), loc='upper left')
plt.tight_layout()
plt.xticks(rotation = 45)
plt.grid(axis='y')

plt.show()
```

```python
country_cost = df.groupby(['Country', 'Sub Category'])['Revenue'].sum().reset_index()

plt.figure(figsize=(13, 8))
sns.barplot(x='Sub Category', y='Revenue', hue='Country', data=country_cost, palette='viridis')

plt.title('Revenue product in each countries')
plt.xlabel('Country')
plt.ylabel('Revenue')

plt.legend(title='Country', bbox_to_anchor=(1.0, 1.2), loc='upper left')
plt.tight_layout()
plt.xticks(rotation = 45)
plt.grid(axis='y')
plt.show()
```

```python
df.groupby('State')['Quantity'].sum().sort_values(ascending = False).head(10).plot(kind='pie', y='Quantity', autopct = '%1.1f%%')
plt.title('Top State by Quantity Distribution')
plt.ylabel('') 
plt.show()
```

```python
df.groupby('State')['Profit'].sum().sort_values(ascending = False).head(10).plot(kind='pie', y='Profit', autopct = '%1.1f%%')
plt.title('Top State by Profit Distribution')
plt.ylabel('') 
plt.show()
```

```python
df.groupby('State')['Profit'].sum().sort_values(ascending = False).tail(10).plot(kind='bar',x = 'State', y='Profit')
plt.title('States by lowest profit')
plt.grid(axis='y')
plt.show()
```

#### Unit price of products should be increased in these Sates.

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
monthly_profit = df['Profit'].resample('ME').sum()
monthly_profit.index = monthly_profit.index.strftime('%Y-%m')
```

```python
plt.figure(figsize=(10, 6))

monthly_profit.plot(kind='bar', color='skyblue')
plt.title('Monthly Profit')
plt.xlabel('Date')
plt.ylabel('Total Profit')
plt.grid(axis='y')
plt.show()
```

### May and June of 2016 made the most profit.
